# Analyse

Betrachtung der vorgeschlagenen Mechanismus:
$$
V := \text{Wert einer Stimme}
$$
$$
P := \text{Voting Power pro Entscheider}
$$
$$
K := \text{Kapitalpool}
$$
$$
n := \text{Zahl der Projekte}
$$
$$
N := \text{Zahl der Teilnehmer} 
$$ 

## kubisches Voting

Kosten $C$ fuer $v$ Votes:

$$
C = \sum_{k=1}^{v}k^2 = \frac{1}{6}(2v^3 + 3v^2 + v)
$$

Zahl der Votes in Abhaengigkeit von $C$ Votingpower:
$$
v(C) = \bigr \lfloor 0.5(\sqrt[3]{0.48075 (1.73205 \sqrt{3888 C^2 - 1} + 108 C} + 0.693361 \sqrt[3]{1.73205 \sqrt{3888 C^2 - 1} + 108 C} - 1) \bigr \rfloor
$$

## Alternative: quadratisches Voting

Kosten $C$ fuer $v$ Votes:

$$
C = \sum_{k=1}^{v}k = \frac{v(v+1)}{2}
$$

Zahl der Votes in Abhaengigkeit von $C$ Votingpower:
$$
v(C) = \bigr \lfloor \frac{1}{2}(\sqrt{8 C + 1} - 1)  \bigr \rfloor
$$


## Annahmen

Maximales Budget eines Entscheiders bei optimalem Verhalten, also gleichmaessiger Verteilung von P auf n:
$$
b_{max} = v(\frac{P}{n}) * n * V
$$


Maximale Anzahl an Entscheidern bei Annahme von $b_{max}$:
$$
N_{max} = \frac{K}{b_{max}}
$$

## Betrachtung von Beispielwerten

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import HTML, display
import tabulate

def cubic_voting(C) -> int:
    ''' assumption: iterative calculation more efficient than formula vor small n '''
    v = 0
    while True:
        C -= (v+1)**2
        if C < 0:
            break
        else:
            v +=1
    return v

def quadratic_voting(C) -> int:
    ''' assumption: iterative calculation more efficient than formula vor small n '''
    v = 0
    while True:
        C -= (v+1)
        if C < 0:
            break
        else:
            v +=1
    return v

def cubic_cost(v) -> int:
    return int(1/6*(2*v**3 + 3*v**2 + v))

def quadratic_cost(v) -> int:
    return int(v*(v + 1) * (1 / 2))

''' CONSTANTS '''
P = 100
V = 50
v = cubic_voting
C = cubic_cost
CUBIC = 'c'
QUADRATIC = 'q'

def build_table(V, P, C, v):
    ''' matrix displaying monetary values no. of people times no. of votes '''
    votes = 1
    no_people = 20
    currency = 'EUR'

    header = ["No. people / No. votes (Voting Power)"]
    while True:
        cost = C(votes)
        if P - cost < 0:
            break
        header.append('{} ({})'.format(votes, cost))
        votes += 1

    table = [header]

    for i in range(1, no_people + 1):
        row = [i]
        for j in range(1, len(header)):
            row.append('{}{}'.format(i * j * V, currency))
        table.append(row)

    print('Example matrix')
    display(HTML(tabulate.tabulate(table, tablefmt='html')))

def adjust_model(mechanism = [('cubic voting', CUBIC), ('quadratic voting', QUADRATIC)], voting_power=P, value_of_vote=V):
    ''' dynamically adjust the model to input '''
    if mechanism == CUBIC:
        v = cubic_voting
        C = cubic_cost
    elif mechanism == QUADRATIC:
        v = quadratic_voting
        C = quadratic_cost
    P = voting_power
    V = value_of_vote
    build_table(V, P, C, v)

print('Adjust model:')
interact(adjust_model, voting_power=(1, 1000), value_of_vote=(1, 200))
print()
```

## Betrachtung von $b_{max}$
Sei $K =10000$ EUR. 
Variieren wir den Wert von V bei $P = 100$:

```python
import matplotlib.pyplot as plt
import numpy as np
import math



def b_max(x, P, V):
    return v(P/x) * x * V

def plot_b_max(plt):
    plt.title('Maximales Budget eines Entscheiders')
    plt.legend(loc='upper left')
    plt.xlabel('n (Zahl der Projekte)')
    plt.ylabel('b_max (EUR)')
    
def plot_N_max(plt):
    plt.title('Maximale Anzahl Entscheider')
    plt.legend(loc='upper left')
    plt.xlabel('n (Zahl der Projekte)')
    plt.ylabel('N_max (Zahl der Entscheider)')
K = 10000

x = np.linspace(1,15,15)

y = np.vectorize(b_max)

# plot the function
plt.plot(x,y(x, 100, 50), 'r', label='V=50 EUR')
plt.plot(x,y(x, 100, 25), 'g', label='V=25 EUR')
plt.plot(x,y(x, 100, 15), 'b', label='V=15 EUR')

plot_b_max(plt)

# show the plot
plt.show()
```

```python
# plot the function
plt.plot(x,K/y(x, 100, 50), 'r', label='V=50 EUR')
plt.plot(x,K/y(x, 100, 25), 'g', label='V=25 EUR')
plt.plot(x,K/y(x, 100, 15), 'b', label='V=15 EUR')

plot_N_max(plt)

# show the plot
plt.show()
```

Variieren wir den Wert von P bei $V = 50$:

```python
# plot the function
plt.plot(x,y(x, 100, 50), 'r', label='P=100')
plt.plot(x,y(x, 50, 50), 'g', label='P=50')
plt.plot(x,y(x, 25, 50), 'b', label='P=25')

plot_b_max(plt)

# show the plot
plt.show()
```

```python
# plot the function
plt.plot(x,K/y(x, 100, 50), 'r', label='P=100')
plt.plot(x,K/y(x, 50, 50), 'g', label='P=50')
plt.plot(x,K/y(x, 25, 50), 'b', label='P=25')

plot_N_max(plt)

# show the plot
plt.show()
```
