# Analyse

Betrachtung der vorgeschlagenen Mechanismus:
$$
V := \text{Wert einer Stimme}
$$
$$
P := \text{Voting Power pro Entscheider}
$$
$$
K := \text{Kapitalpool}
$$
$$
n := \text{Zahl der Projekte}
$$
$$
N := \text{Zahl der Teilnehmer} 
$$ 


Maximales Budget eines Entscheiders bei optimalem Verhalten, also gleichmaessiger Verteilung von P auf n:
$$
b_{max} = \sqrt{\frac{P}{n}} * n * V
$$


Maximale Anzahl an Entscheidern bei Annahme von $b_{max}$:
$$
N_{max} = \frac{K}{b_{max}}
$$

$$
C = \sum_{k=1}^{n}k^2
$$

## Betrachtung von $b_{max}$
Sei $K =10000$ EUR. 
Variieren wir den Wert von V bei $P = 100$:

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def b_max(x, P, V):
    return math.sqrt(P/x) * x * V

def plot_b_max(plt):
    plt.title('Maximales Budget eines Entscheiders')
    plt.legend(loc='upper left')
    plt.xlabel('n (Zahl der Projekte)')
    plt.ylabel('b_max (EUR)')
    
def plot_N_max(plt):
    plt.title('Maximale Anzahl Entscheider')
    plt.legend(loc='upper left')
    plt.xlabel('n (Zahl der Projekte)')
    plt.ylabel('N_max (Zahl der Entscheider)')
K = 10000

x = np.linspace(1,15,15)

y = np.vectorize(b_max)

# plot the function
plt.plot(x,y(x, 100, 50), 'r', label='V=50 EUR')
plt.plot(x,y(x, 100, 25), 'g', label='V=25 EUR')
plt.plot(x,y(x, 100, 15), 'b', label='V=15 EUR')

plot_b_max(plt)

# show the plot
plt.show()
```

```python
# plot the function
plt.plot(x,K/y(x, 100, 50), 'r', label='V=50 EUR')
plt.plot(x,K/y(x, 100, 25), 'g', label='V=25 EUR')
plt.plot(x,K/y(x, 100, 15), 'b', label='V=15 EUR')

plot_N_max(plt)

# show the plot
plt.show()
```

Variieren wir den Wert von P bei $V = 50$:

```python
# plot the function
plt.plot(x,y(x, 100, 50), 'r', label='P=100')
plt.plot(x,y(x, 50, 50), 'g', label='P=50')
plt.plot(x,y(x, 25, 50), 'b', label='P=25')

plot_b_max(plt)

# show the plot
plt.show()
```

```python
# plot the function
plt.plot(x,K/y(x, 100, 50), 'r', label='P=100')
plt.plot(x,K/y(x, 50, 50), 'g', label='P=50')
plt.plot(x,K/y(x, 25, 50), 'b', label='P=25')

plot_N_max(plt)

# show the plot
plt.show()
```
